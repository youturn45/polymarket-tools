```python
import subprocess

def encrypt_with_ssh(data, output_file):
    subprocess.run(
        ["age", "-R", str(Path("~/.ssh/Youturn.pub").expanduser()), 
         "-o", output_file],
        input=data,
        check=True
    )

def decrypt_with_ssh(encrypted_file):
    result = subprocess.run(
        ["age", "-d", "-i", str(Path("~/.ssh/Youturn.pub").expanduser()),
         encrypted_file],
        capture_output=True,
        check=True
    )
    return result.stdout
```
